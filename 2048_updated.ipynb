```python
import random
l1=[" "]*16

#Generates numbers in random places.
l1[random.choice(list(range(0,7)))]=2
l1[random.choice(list(range(8,15)))]=2
cnt=5
abcd=1

#Defining a function to print 2048 table.

def print_2048():
    print(" "+"------ "*4)
    for i in range(16):
        print("|",str(l1[i]).center(5," "),end="")
        if (1+i)%4==0:
            print("|")
            print(" "+"------ "*4)
print_2048()

#Defining a function that applies changes to 2048 table based on user's choice.

def choice_apply(value,index):
    
    i=index
    i1=0
    i2=0
    i3=0
    if value=='r':
        i1=i-1
        i2=i-2
        i3=i-3
    elif value=='l':
        i1=i+1
        i2=i+2
        i3=i+3
    elif value=='t':
        i1=i+4
        i2=i+8
        i3=i+12
    elif value=='d':
        i1=i-4
        i2=i-8
        i3=i-12
        
    if(l1[i]==" " and l1[i1]==" " and l1[i2]==" "):
        l1[i]=l1[i3]
        l1[i3]=" "
                
    elif(l1[i]==" " and l1[i1]==" "):
        l1[i]=l1[i2]
        l1[i1]=l1[i3]
        l1[i2]=" "
        l1[i3]=" "
                
    elif(l1[i2]==" " and l1[i1]==" "):
        l1[i1]=l1[i3]
        l1[i2]=" "
        l1[i3]=" "

    elif(l1[i]==" " and l1[i2]==" "):
        l1[i]=l1[i1]
        l1[i1]=l1[i3]
        l1[i2]=" "
        l1[i3]=" "
                
    elif(l1[i1]==" " and l1[i3]==" "):
        l1[i1]=l1[i2]
        l1[i2]=" "
        l1[i3]=" "
                
    elif(l1[i]==" "):
        l1[i]=l1[i1]
        l1[i1]=l1[i2]
        l1[i2]=l1[i3]
        l1[i3]=" "
               
    elif(l1[i1]==" " ):
        l1[i1]=l1[i2]
        l1[i2]=l1[i3]
        l1[i3]=" "
                
    elif(l1[i2]==" " ):
        l1[i2]=l1[i3]
        l1[i3]=" "
            
    elif(l1[i3]==" " ):
        l1[i3]=" "
                
    else:
        gen=1
                
    if l1[i]==l1[i1] and l1[i]!=' ':
        l1[i]=l1[i]+l1[i1]
        l1[i1]=l1[i2]
        l1[i2]=l1[i3]
        l1[i3]=" "

    elif l1[i1]==l1[i2]and l1[i1]!=' ':
        if l1[i]==' ':
            l1[i]=l1[i1]+l1[i2]
            l1[i1]=l1[i3]
            l1[i2]=" "
            l1[i3]=" "

        else:
            l1[i1]=l1[i1]+l1[i2]
            l1[i2]=l1[i3]
            l1[i3]=" "

    elif l1[i2]==l1[i3] and l1[i2]!=' ':
        if l1[i]!=' ' and l1[i1]==' ':
            l1[i]=l1[i2]+l1[i3]
            l1[i1]=" "
            l1[i2]=" "
            l1[i3]=" "

        elif l1[i1]==' ':
            l1[i]=l1[i2]+l1[i3]
            l1[i2]=" "
            l1[i3]=" "

        else:
            l1[i2]=l1[i2]+l1[i3]
            l1[i3]=" "

#Repeating loop until game over.
            
while l1.count(" ")>0 or cnt!=0:
    gen=0

    in1=input("Select one:\nr-Right\nl-Left\nt-Top\nd-Dowm\n->") # Taking input from user.
    
    #Performing action based on user input.
    if(in1=="r"):
        for i in [3,7,11,15]:
            choice_apply(in1,i)
                
    elif(in1=="l"):
        for i in [0,4,8,12]:
            choice_apply(in1,i)
            
    elif(in1=="t"):
        for i in [0,1,2,3]:
            choice_apply(in1,i)

    elif(in1=="d"):
        for i in [12,13,14,15]:
            choice_apply(in1,i)

    else:
        print("Please enter from list")
    
    
    if in1 in ["r","l","t","d"]:
        l2=[]
        #Preparing a list of cells which are blank.
        for j in range(16):
            if l1[j]==" ":
                l2.append(j)
        #Generating 2 or 4 random number in blank cell.        
        if(len(l2)!=0) and gen==0:
            l1[random.choice(l2)]=random.choice([2,4,2,2,2])
            
        print_2048() #Printing 2048 table.
    
    if(l1[0]==l1[1] or l1[1]==l1[2] or l1[2]==l1[3] or l1[4]==l1[5] or l1[5]==l1[6] or l1[6]==l1[7] or l1[8]==l1[9] or l1[9]==l1[10] or l1[10]==l1[11] or l1[12]==l1[13] or l1[13]==l1[14] or l1[14]==l1[15] or l1[0]==l1[4] or l1[4]==l1[8] or l1[8]==l1[12] or l1[1]==l1[5] or l1[5]==l1[9] or l1[9]==l1[13] or l1[2]==l1[6] or l1[6]==l1[10] or l1[10]==l1[14] or l1[3]==l1[7] or l1[7]==l1[11] or l1[11]==l1[15]):
        cnt=1
    else:
        cnt=0
    if 2048 in l1 and abcd==1:
        input("Yeah you reached 2048!\n:-)\npress any key to continue:") #Printing if user reached 2048.
        abcd=0
    
print("Game Over!") # printing 'game over' if there is no chace to continue the game
```
