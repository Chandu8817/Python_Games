```python
values=[[0,0,0,0,0,0,6,8,0],
       [0,0,0,0,7,3,0,0,9],
       [3,0,9,0,0,0,0,4,5],
       [4,9,0,0,0,0,0,0,0],
       [8,0,3,0,5,0,9,0,2],
       [0,0,0,0,0,0,0,3,6],
       [9,6,0,0,0,0,3,0,8],
       [7,0,0,6,8,0,0,0,0],
       [0,2,8,0,0,0,0,0,0]]

values=[[9,0,1,0,8,5,0,0,4],
        [0,7,4,1,0,3,8,0,0],
        [0,0,0,0,4,6,0,7,0],
        [7,9,8,4,0,1,6,2,0],
        [0,2,3,9,0,0,5,0,7],
        [5,1,6,0,0,2,0,0,0],
        [0,3,0,8,2,4,0,6,5],
        [0,0,9,0,1,0,3,0,2],
        [6,0,0,5,3,9,7,8,0]]

#Defining a function to priny sudoku table.
def print_sudoku_table(values):
    print('|---'*9+'|')
    for i in values:
        print('| '+" | ".join(list(map(str,i)))+' |')
        print('|---'*9+'|')
        
print_sudoku_table(values) #Calling sudoku table function.
vals_row=[]
cnt=0

# Repeate loop until we get Solution
while True:
    n=cnt
    cnt=0
    for i in values:
        cnt=cnt+i.count(0)#Finding count of cells which are empty
    if n==cnt:
        if cnt!=0:
            print('Puzzle has multiple solutions......cant solve')
        break
    cols_box=[]
    vals_col=[]
    for i in range(9):#Repeat for each row.
        vals_row=[]
        vals_row1=[]
        rows_box=[]
        
        for j in range(9):#Repeat for each column.
            
            if values[i][j]==0:# Processing the cells which are blank.
                b=[]
                r_b=[]
                c_b=[]
                box=[]
                r_box1=[]
                r_box2=[]
                c_box1=[]
                c_box2=[]
                
                #Dividing sudoku table into different boxes.
                
                b1=[x1 for x in values[:3] for x1 in x[:3]];
                b2=[x1 for x in values[:3] for x1 in x[3:6]];
                b3=[x1 for x in values[:3] for x1 in x[6:9]];
                b4=[x1 for x in values[3:6] for x1 in x[:3]];
                b5=[x1 for x in values[3:6] for x1 in x[3:6]];
                b6=[x1 for x in values[3:6] for x1 in x[6:9]];
                b7=[x1 for x in values[6:] for x1 in x[:3]];
                b8=[x1 for x in values[6:] for x1 in x[3:6]];
                b9=[x1 for x in values[6:] for x1 in x[6:9]];
                
                
                #Defining functon to get current box information(possible elements to be inserted in blank cells).
                def box_values(r_from,r_to,c_from,c_to,b):
                    box_values=[]
                    for r_v in range(r_from,r_to):
                            for c_v in range(c_from,c_to):   
                                if values[r_v][c_v]==0:
                                    box_values.append([r_v,c_v,[x for x in range(1,10) if x not in values[r_v] and x not in [v[c_v] for v in values ] and x not in b]])
                    return box_values
                
                #Calling box_values function for each box.
                
                box1=box_values(0,3,0,3,b1)
                box2=box_values(0,3,3,6,b2)
                box3=box_values(0,3,6,9,b3)
                box4=box_values(3,6,0,3,b4)
                box5=box_values(3,6,3,6,b5)
                box6=box_values(3,6,6,9,b6)
                box7=box_values(6,9,0,3,b7)
                box8=box_values(6,9,3,6,b8)
                box9=box_values(6,9,6,9,b9)
            
#Getting information of other boxes which are in same row or column of a current cell.
                
                if i in [0,1,2] and j in [0,1,2]:
                    b=b1
                    r_b=set(b2)&set(b3)
                    c_b=set(b4)&set(b7)
                    box=box1
                    r_box1=box2
                    r_box2=box3
                    c_box1=box4
                    c_box2=box7
                    
                elif i in [0,1,2] and j in [3,4,5]:
                    b=b2
                    r_b=set(b1)&set(b3)
                    c_b=set(b5)&set(b8)
                    box=box2
                    r_box1=box1
                    r_box2=box3
                    c_box1=box5
                    c_box2=box8
                    
                elif i in [0,1,2] and j in [6,7,8]:
                    b=b3
                    r_b=set(b2)&set(b1)
                    c_b=set(b6)&set(b9)
                    box=box3
                    r_box1=box1
                    r_box2=box2
                    c_box1=box6
                    c_box2=box9
                elif i in [3,4,5] and j in [0,1,2]:
                    b=b4
                    r_b=set(b5)&set(b6)
                    c_b=set(b1)&set(b7)
                    box=box4
                    r_box1=box5
                    r_box2=box6
                    c_box1=box1
                    c_box2=box7
                    
                elif i in [3,4,5] and j in [3,4,5]:
                    b=b5
                    r_b=set(b4)&set(b6)
                    c_b=set(b2)&set(b8)
                    box=box5
                    r_box1=box4
                    r_box2=box6
                    c_box1=box2
                    c_box2=box8
                elif i in [3,4,5] and j in [6,7,8]:
                    b=b6
                    r_b=set(b5)&set(b4)
                    c_b=set(b3)&set(b9)
                    box=box6
                    r_box1=box4
                    r_box2=box5
                    c_box1=box3
                    c_box2=box9
                elif i in [6,7,8] and j in [0,1,2]:
                    b=b7
                    r_b=set(b8)&set(b9)
                    c_b=set(b4)&set(b1)
                    box=box7
                    r_box1=box8
                    r_box2=box9
                    c_box1=box1
                    c_box2=box4
                elif i in [6,7,8] and j in [3,4,5]:
                    b=b8
                    r_b=set(b7)&set(b9)
                    c_b=set(b2)&set(b5)
                    box=box8
                    r_box1=box7
                    r_box2=box9
                    c_box1=box2
                    c_box2=box5
                elif i in [6,7,8] and j in [6,7,8]:
                    b=b9
                    r_b=set(b7)&set(b8)
                    c_b=set(b3)&set(b6)
                    box=box9
                    r_box1=box7
                    r_box2=box8
                    c_box1=box3
                    c_box2=box6
                 
                # l stores all possible values for  current cell.
                # l has all Values that are not present in current row,current column and current box.
                l=[x for x in range(1,10) if x not in values[i] and x not in [v[j] for v in values ] and x not in b]
                
                box_1=[]
                for vv in box:
                    for vv1 in vv[2]:
                        box_1.append(vv1)
                
                #Placing value in a empty cell if that is only possible cell for that value.
                
                for vv in box_1:
                    if box_1.count(vv)==1 and vv not in b:
                        for v1 in box:
                            if vv in v1[2]:
                                print('box:',v1[0],v1[1],vv)
                                values[v1[0]][v1[1]]=vv
                                print_sudoku_table(values)
                                break
                 
                #Defining function for eliminating values from l.
                
                def remove_from_l(other_box,r_c,ind):
                    for val_table in range(1,10):
                        c=0
                        p=0
                        if val_table in l:
                            for val in other_box:
                                if ind!= val[r_c] and val_table in val[2]:
                                    c+=1
                                if ind==val[r_c] and val_table in val[2]:
                                    p+=1
                            if c==0 and p!=0:
                                l.remove(val_table)
                                
                #Calling frunction remove_from_l.                
                remove_from_l(r_box1,0,i)
                remove_from_l(r_box2,0,i)
                remove_from_l(c_box1,1,j)
                remove_from_l(c_box2,1,j)

                for item in l:
                    vals_row.append(item)#Appending each avlue of l to vals_row variable.
                    
                vals_col.append([j,i,l])# Appending column,row,l
                vals_row1.append([i,j,l])# Appending row,column,l
                                
                #Placing value in a empty cell if it has ony one possible number.
                if len(l)==1:
                    if values[i][j]==0:
                        print('l:',i,j,l)
                        values[i][j]=l[0]
                        print_sudoku_table(values)

                if len(list(r_b&set(l)))!=0:
                    rows_box.append([(i,j),list(r_b&set(l))])
                if len(list(c_b&set(l)))!=0:
                    cols_box.append([(j,i),list(c_b&set(l))])
                    
        #Placing value in a empty cell if that is only possible cell for that value in the current row.
        for val in vals_row:
            if vals_row.count(val)==1:
                for val1 in vals_row1:
                    if val in val1[2] and val not in values[val1[0]]:
                        if values[val1[0]][val1[1]]==0:
                            print('row',val1[0],val1[1],val)
                            values[val1[0]][val1[1]]=val
                            print_sudoku_table(values)
        
        rr=[]
        cc=[]
        
        #Placing value in a empty cell if that only possible cell in current box.
        for r in rows_box:
            if len(r[1])==1 and vals_row.count(r[1][0])==1 and  values[r[0][0]][r[0][1]]==0:
                if values[r[0][0]][r[0][1]]==0:
                    print('r:',r[0][0],r[0][1],r[1][0])
                    values[r[0][0]][r[0][1]]=r[1][0]
                    print_sudoku_table(values)

    #Placing value in a empty cell if that only possible cell in current box
    for c in cols_box:
        cc=[]
        for cc_item in [x[2] for x in vals_col if x[0]==c[0][0]]:
            for c_i in cc_item:
                cc.append(c_i)
        if len(c[1])==1 and cc.count(c[1][0])==1:
            if values[c[0][1]][c[0][0]]==0:
                print('c:',c[0][1],c[0][0],c[1][0])
                values[c[0][1]][c[0][0]]=c[1][0]
                print_sudoku_table(values)

#Printing final table
print_sudoku_table(values)
```
